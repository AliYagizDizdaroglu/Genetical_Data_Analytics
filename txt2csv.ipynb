```python
import pandas as pd

# data reading
df = pd.read_csv('ML Engineer Candidate Prompt Dataset.txt', delimiter = "\t")
print(df.head())
print(df.shape)
```

```python
# data foreplay

df_transpose = df.T
list_of_feats = list(df_transpose.iloc[0,:])
#print(list_of_feats[:10])
df_transpose.columns = list_of_feats
df_transpose = df_transpose.iloc[1:,:]
df_transpose = df_transpose.reset_index()
df_transpose['initial weight'] = df_transpose.iloc[:,0]
df_transpose.rename(columns={'index': 'initial weight'}, inplace=True, errors='raise')
#print(df_transpose.head())
#print(df_transpose.shape)
df_transpose = df_transpose.iloc[:,:-1]
print(df_transpose.head())

# regularized df
df = df_transpose
```

```python
lst_age = df['age_at_diagnosis'].tolist()
lst_days = df['days_to_birth'].tolist()
lst_days = [str(abs(int(elem))) for elem in lst_days]

lst_age == lst_days # True, might drop it later

df
```

```python
cat_features = ['paper_clinical_stage','demographic_id','vital_status']

for col in df.columns:
    if col not in cat_features:
        df[col] = df[col].astype(float)
               
df.iloc[:,:20].info()
```

```python
df.iloc[:,:20].info()
df.to_csv('data.csv', index = True, encoding='utf-8')
```

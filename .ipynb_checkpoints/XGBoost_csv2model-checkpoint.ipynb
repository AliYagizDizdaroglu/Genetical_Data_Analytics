```python
import pandas as pd

# data reading
df_ = pd.read_csv('data.csv')
df = df_
```

```python
# dropping demographic_id as it is unique to person and no use for us
df = df.drop(['demographic_id'], axis=1)

# dropping age_at_diagnosis, we have the same as days_to_birth
df = df.drop(['age_at_diagnosis'], axis=1)

df.info()
```

```python
df
```

```python
# data ready to go
df = df.iloc[:,1:]

# changing paper_clinical_stage to number respectively (1,2,3,4)
df.loc[df['paper_clinical_stage'] == 'Stage I', 'paper_clinical_stage'] = 1
df.loc[df['paper_clinical_stage'] == 'Stage II', 'paper_clinical_stage'] = 2
df.loc[df['paper_clinical_stage'] == 'Stage III', 'paper_clinical_stage'] = 3
df.loc[df['paper_clinical_stage'] == 'Stage IV', 'paper_clinical_stage'] = 4

# changing vital_status to number respectively (0,1)
df.loc[df['vital_status'] == 'Dead', 'vital_status'] = 0
df.loc[df['vital_status'] == 'Alive', 'vital_status'] = 1

# finding current age
df = df.assign(days_to_birth = df['days_to_birth'] / -365.24)
df.rename(columns={'days_to_birth': 'age_current'}, inplace=True, errors='raise')
df
```

```python
# had to convert to int indirectly
df["vital_status"] = df["vital_status"].astype(float).astype(int)
df["paper_clinical_stage"] = df["paper_clinical_stage"].astype(float).astype(int)

# dropping dead with unknown year_of_death as it makes no sense
df = df.drop(df[(df['vital_status'] == 0) & (df['year_of_death'].isnull())].index)
df
```

```python
df.insert(5, 'age_at_death', df['year_of_death'] - (df['year_of_birth']))
```

```python
df.insert(3, 'years_in_diagnosis', df['age_current'] - df['age_at_index'])
```

```python
df
```

```python
# df_diagnosed_dead will be used for day calculation later
df_diagnosed_dead = df
df_diagnosed_dead.columns
```

```python
# dropping death related columns to prevent data leak
df = df.drop(['age_at_death'], axis=1)
df = df.drop(['year_of_death'], axis=1)
```

```python
#df = df.drop(['days_to_death'], axis=1)
```

```python
# implementing the XGBoost model
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = df.loc[:, df.columns != 'vital_status'], df['vital_status']
print(X)

# split data into train and test sets
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state = 1)

print(X_train)

# fit model on training data
model = XGBClassifier()
model.fit(X_train, y_train)

print(model)

xgb.plot_importance(model)

from matplotlib import pyplot as plt
plt.figure(figsize=(20,20))
xgb.plot_importance(model,max_num_features=20)
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
